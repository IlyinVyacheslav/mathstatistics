##### Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics
from scipy import stats
from tabulate import tabulate
```

##### Read dataset

```python
data = pd.read_csv("cars93.csv")
```

##### What types are represented in the dataset?

```python
print(data["Type"].unique())

types_counts = data["Type"].value_counts()
print(types_counts)
```

##### Which type is the most common?

```python
print(types_counts.idxmax(), ':', types_counts.max())
```

##### Which type is the least common?

```python
print(types_counts.idxmin(), ':', types_counts.min())
```

```python
diff_price_data = data["Max.Price"] - data["Min.Price"]
```

##### Sample mean

```python
print(statistics.mean(diff_price_data))
```

##### Sample variance

```python
print(statistics.variance(diff_price_data))
```

##### Median

```python
print(statistics.median(diff_price_data))
```

##### Quantiles

```python
print(statistics.quantiles(diff_price_data))
```

##### Auxiliary functions

```python
def sort_and_print_list(list):
    for item in sorted(list): 
        print(item)
```

```python
def print_graphics(data):
    res = stats.ecdf(data)
    res.cdf.plot(plt.subplot())
    plt.title("Graph of ecdf")
    plt.show()

    plt.hist(data)
    plt.title("Histogram")
    plt.show()

    plt.boxplot(data)
    plt.title("Box-plot")
    plt.show()
```

##### Draw graphs of whole set

```python
print_graphics(diff_price_data)
```

##### Consider each type of car separately

```python
data_van     = data[data["Type"] == 'Van']
data_large   = data[data["Type"] == 'Large']
data_sporty  = data[data["Type"] == 'Sporty']
data_compact = data[data["Type"] == 'Compact']
data_small   = data[data["Type"] == 'Small']
data_midsize = data[data["Type"] == 'Midsize']

diff_price_data_van     = data_van["Max.Price"] - data_van["Min.Price"]
diff_price_data_large   = data_large["Max.Price"] - data_large["Min.Price"]
diff_price_data_sporty  = data_sporty["Max.Price"] - data_sporty["Min.Price"]
diff_price_data_compact = data_compact["Max.Price"] - data_compact["Min.Price"]
diff_price_data_small   = data_small["Max.Price"] - data_small["Min.Price"]
diff_price_data_midsize = data_midsize["Max.Price"] - data_midsize["Min.Price"]
```

```python
head = ["", "mean", "variance", "median", "quantiles"]
mydata = [
    [
        "All",
        statistics.mean(diff_price_data),
        statistics.variance(diff_price_data),
        statistics.median(diff_price_data),
        statistics.quantiles(diff_price_data)
    ],
    [
        "Van",
        statistics.mean(diff_price_data_van),
        statistics.variance(diff_price_data_van),
        statistics.median(diff_price_data_van),
        statistics.quantiles(diff_price_data_van)
    ],
    [
        "Large",
        statistics.mean(diff_price_data_large),
        statistics.variance(diff_price_data_large),
        statistics.median(diff_price_data_large),
        statistics.quantiles(diff_price_data_large)
    ],
    [
        "Sporty",
        statistics.mean(diff_price_data_sporty),
        statistics.variance(diff_price_data_sporty),
        statistics.median(diff_price_data_sporty),
        statistics.quantiles(diff_price_data_sporty)
    ],
    [
        "Compact",
        statistics.mean(diff_price_data_compact),
        statistics.variance(diff_price_data_compact),
        statistics.median(diff_price_data_compact),
        statistics.quantiles(diff_price_data_compact)
    ],
    [
        "Small",
        statistics.mean(diff_price_data_small),
        statistics.variance(diff_price_data_small),
        statistics.median(diff_price_data_small),
        statistics.quantiles(diff_price_data_small)
    ],
    [
        "Midsize",
        statistics.mean(diff_price_data_midsize),
        statistics.variance(diff_price_data_midsize),
        statistics.median(diff_price_data_midsize),
        statistics.quantiles(diff_price_data_midsize)
    ],
]
print(tabulate(mydata, headers=head, tablefmt="grid"))
```
